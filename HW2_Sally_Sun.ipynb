```python
def get_business(query,api_key,min_rating,max_results):
    import requests
    url="https://maps.googleapis.com/maps/api/place/textsearch/json?key=%s&input=%s"  % (api_key,query)
    response = requests.get(url)
    if not response.status_code == 200:
        print("HTTP error",response.status_code)
   
    output=list()
    for item in response.json()['results']: 
        if item['rating'] >= min_rating:
            try:
                output.append(tuple([item['name'],item['formatted_address'],item['opening_hours'],item['price_level'],item['rating']]))
            except:
                output.append(tuple([item['name'],item['formatted_address'],item['opening_hours'],None,item['rating']]))
            if len(output)>= max_results:
                return output

    while True:   
        if 'next_page_token' not in response.json().keys():
            return output
        else:
            new_url="https://maps.googleapis.com/maps/api/place/textsearch/json?pagetoken=%s&key=%s" % (response.json()['next_page_token'],api_key)
            response = requests.get(new_url)
            while response.json().get('status') != 'OK':
                from time import sleep
                from random import random
                sleep(random())
                response = requests.get(new_url)
            
            for item in response.json()['results']: 
                if item['rating'] >= min_rating:
                    try:
                        output.append(tuple([item['name'],item['formatted_address'],item['opening_hours'],item['price_level'],item['rating']]))
                    except:
                        output.append(tuple([item['name'],item['formatted_address'],item['opening_hours'],None,item['rating']]))
                    if len(output)>=max_results:
                        return output
```
